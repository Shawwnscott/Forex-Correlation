# Clean Data

#### GBPUSD

```python
gbp_usd.isnull().sum()
```

```python
gbp_usd = gbp_usd.ffill()
```

```python
gbp_usd = gbp_usd.resample("D", label = "right").last()
gbp_usd.head()
```

```python
gbp_usd.isnull().sum()
```

```python
gbp_usd = gbp_usd.ffill()
gbp_usd
```

```python
gbp_usd = pd.DataFrame(gbp_usd['Close'])
gbp_usd
```

```python
gbp_usd = gbp_usd.rename(columns = {"Close":"GBPUSD"})
gbp_usd
```

#### Merge Tables

```python
forex = pd.merge(pd.merge(eur_usd,gbp_usd,left_index=True, right_index=True),usd_cad,left_index=True, right_index=True)
```

```python
forex = pd.merge(forex,aud_usd, left_index=True, right_index=True)
```

```python
forex = pd.merge(forex,usd_jpy, left_index=True, right_index=True)
forex
```

```python
forex.to_csv('EURUSDx Correlation 2010-2020.csv')
```

## Relationship Analysis

```python
correlation = forex.corr()
```

```python
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns, annot = True)
```
